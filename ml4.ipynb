```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
iris_data = load_iris()
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris_df['target'] = iris_data.target
```

```python
# Display iris dataset
print(iris_df.head())
```

```python
X = iris_df.drop('target', axis=1)
y = iris_df['target']
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Initialize Decision Tree Classifier
clf =  DecisionTreeClassifier(criterion="gini",random_state=100, max_depth=5, min_samples_leaf=8)
```

```python
# Train the classifier
clf.fit(X_train, y_train)
```

```python
# Plot the decision tree
plt.figure(figsize=(12,8))
tree.plot_tree(clf, feature_names=iris_data.feature_names, class_names=iris_data.target_names, filled=True)
plt.show()
```

```python
# Predict the accuracy using Decision Tree Classifier
accuracy = clf.score(X_test, y_test)
print("Accuracy:", accuracy)
```
